## Question_1st:- Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing?

```python
# The z-statistic (also called the z-score) is a measure that describes how many standard deviations a data point is from the mean of a distribution. It's calculated as:


#                                        z= (X−μ)/σ

# where:
     # X is the data point or sample mean,
    #  μ is the population mean,
     # σ is the population standard deviation.
# Relationship to the Standard Normal Distribution:-
    # The z-statistic corresponds to the standard normal distribution (also known as the z-distribution), which is a normal distribution with a mean of 0 and a standard deviation of 1. When you standardize a value using the z-statistic formula, you are effectively transforming it into this standard normal form.

# How is the Z-Statistic Used in Hypothesis Testing?
    # In hypothesis testing, the z-statistic is used to determine how far a sample mean deviates from the population mean under the null hypothesis. It helps in calculating the p-value, which tells us the probability of observing a result as extreme as the test statistic under the assumption that the null hypothesis is true.

# Here’s how it is used in hypothesis testing:

    # 1. Formulate the hypotheses:

          # Null hypothesis (H₀): The population mean is equal to some value (e.g.,μ=μ0).
        
          # Alternative hypothesis (H₁): The population mean is different from the null hypothesis value.
            
    # 2.Calculate the z-statistic:
    
          # Using the sample data, compute the z-statistic to see how far the sample mean is from the hypothesized population mean.

    # 3.Find the p-value:
    
          # The p-value is determined by looking up the z-statistic in the standard normal distribution table (or using software). This gives the probability of obtaining a result as extreme as the observed one.

   # 4.Make a decision:
         # if the p-value is less than a chosen significance level (e.g., α=0.05), reject the null hypothesis. Otherwise, fail to reject it.
    
    
import scipy.stats as stats
import numpy as np

# Hypothetical data
population_mean = 100
population_std = 15
sample_mean = 105
sample_size = 30
alpha = 0.05

# Z-statistic formula
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print the results
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

## Question_2nd:-: What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)?

```python
# P-Value Definition:-

      # The p-value is the probability of observing results as extreme as (or more extreme than) the data observed, assuming the null hypothesis is true. 
      # It helps determine the strength of evidence against the null hypothesis.

# P-Value in Hypothesis Testing

      # If the p-value is less than the significance level (α), typically 0.05, we reject the null hypothesis. 
      # This suggests the observed data is unlikely under the null hypothesis.
        
      # If the p-value is greater than α, we fail to reject the null hypothesis, implying the data is consistent with the null hypothesis.
    
# Interpretation of a Small P-Value (e.g., 0.01)
      # A small p-value (e.g., 0.01) means:

          # There is strong evidence against the null hypothesis.
          # The probability of observing such data by chance is very low (1%).





import scipy.stats as stats
import numpy as np

# Step 1: Hypothetical data
population_mean = 100
population_std = 15
sample_mean = 105
sample_size = 30
alpha = 0.05  # Step 2: Significance level

# Step 3: Z-statistic calculation
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 4: P-value calculation (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 5: Print results
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

## Question_3rd:-Compare and contrast the binomial and Bernoulli distributions. 

```python
# Bernoulli Distribution:-
      # Definition: A Bernoulli distribution represents a single trial with two possible outcomes: success (1) or failure (0).
    
      # Parameters:
             # p: Probability of success (0 ≤ p ≤ 1).
     # Use Case: It is used when modeling a single experiment, like flipping a coin (heads or tails).
    
# Binomial Distribution:-
     # Definition: A Binomial distribution represents the number of successes in n independent Bernoulli trials.
    
     # Parameters:
            # n: Number of trials.
            # p: Probability of success on each trial.
    # Use Case: It is used for multiple experiments, like flipping a coin n times and counting how many heads you get.
    
# Comparison:-
    # Nature:Bernoulli is a single trial; Binomial is multiple trials.
    
    # Parameters: Bernoulli has one parameter p; Binomial has two parameters n and p.
    
    # Outcomes: Bernoulli outcomes are just 0 or 1; Binomial outcomes can range from 0 to n.
    
    
    
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom

# Step 1: Define parameters for the distributions
p = 0.5  # Probability of success
n = 10   # Number of trials for binomial

# Step 2: Create a range of x values for the binomial distribution
x_binom = np.arange(0, n + 1)

# Step 3: Calculate the Bernoulli distribution for one trial
bernoulli_prob = bernoulli.pmf(1, p)
bernoulli_prob_0 = bernoulli.pmf(0, p)

# Step 4: Calculate the Binomial distribution probabilities
binomial_prob = binom.pmf(x_binom, n, p)

# Step 5: Plot the distributions
plt.figure(figsize=(12, 6))

# Bernoulli Distribution Plot
plt.subplot(1, 2, 1)
plt.bar([0, 1], [bernoulli_prob_0, bernoulli_prob], color=['blue', 'orange'])
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.title('Bernoulli Distribution')
plt.xlabel('Outcomes')
plt.ylabel('Probability')

# Binomial Distribution Plot
plt.subplot(1, 2, 2)
plt.bar(x_binom, binomial_prob, color='green')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

# Show the plots
plt.tight_layout()
plt.show()
```

## Question_4th:-: Under what conditions is the binomial distribution used, and how does it relate to the Bernoulli distribution?

```python
# Conditions for Using Binomial Distribution:-

     # 1.Fixed Number of Trials (n): The number of trials is predetermined and constant.
    
     # 2.Two Possible Outcomes: Each trial results in either success (1) or failure (0).
        
     # 3.Independent Trials: The outcome of one trial does not affect the others.
    
     # 4.Constant Probability (p): The probability of success remains the same across all trials.
        
# Relation to Bernoulli Distribution:-

     # Each individual trial in a Binomial distribution is a Bernoulli trial.
    
     # The Binomial distribution can be viewed as the sum of multiple independent Bernoulli trials, where each trial has the same success probability p.
        
        
        
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom

# Step 1: Define parameters
p = 0.6  # Probability of success
n = 5    # Number of trials

# Step 2: Bernoulli Distribution for a single trial
bernoulli_success = bernoulli.pmf(1, p)
bernoulli_failure = bernoulli.pmf(0, p)

# Step 3: Binomial Distribution probabilities
x_binom = np.arange(0, n + 1)  # Possible successes from 0 to n
binomial_probs = binom.pmf(x_binom, n, p)

# Step 4: Print Bernoulli outcomes
print(f'Bernoulli Distribution: Success (1) Probability: {bernoulli_success:.2f}, Failure (0) Probability: {bernoulli_failure:.2f}')

# Step 5: Print Binomial probabilities
print("\nBinomial Distribution Probabilities:")
for i, prob in enumerate(binomial_probs):
    print(f'Probability of {i} successes: {prob:.2f}')

# Step 6: Plotting the distributions
plt.figure(figsize=(10, 5))

# Bernoulli Distribution Plot
plt.subplot(1, 2, 1)
plt.bar([0, 1], [bernoulli_failure, bernoulli_success], color=['blue', 'orange'])
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.title('Bernoulli Distribution')
plt.ylabel('Probability')

# Binomial Distribution Plot
plt.subplot(1, 2, 2)
plt.bar(x_binom, binomial_probs, color='green')
plt.title('Binomial Distribution (n=5, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

# Show the plots
plt.tight_layout()
plt.show()
```

## Question_5th:-: What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution?

```python
# The Poisson distribution is a probability distribution that is useful in various scenarios, 
  # particularly in counting events that occur independently within a fixed interval of time or space. 
  # Here are its key properties and appropriate usage conditions:


# Key Properties of the Poisson Distribution

      # 1.Discrete Distribution: It models the number of events occurring in a fixed interval.
    
      # 2.Parameter: It is characterized by a single parameter λ (lambda), which is the average rate of occurrence of the event.
        
      # 3.Independence: Events occur independently of each other.
    
     # 4.Probability Mass Function (PMF):  P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}


             # where 

                   # k is the number of occurrences, 

                   # e is Euler's number, and 

                   # λ is the average rate.
# Mean and Variance: Both the mean and variance of a Poisson distribution are equal to λ.
    
    
    
# When to Use the Poisson Distribution:-

     # Counting Events: When you want to model the number of events that occur in a fixed interval (time, area, etc.).
     # Rare Events: It is suitable for events that happen infrequently.
     # Independent Events: When the occurrence of one event does not affect the occurrence of another.
    
    
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define the parameter (lambda)
lambda_param = 3  # Average number of events in the interval

# Step 2: Create a range of x values (number of events)
x_values = np.arange(0, 11)  # Counting events from 0 to 10

# Step 3: Calculate the Poisson probabilities
poisson_probs = poisson.pmf(x_values, lambda_param)

# Step 4: Print the probabilities
print("Poisson Distribution Probabilities:")
for k, prob in zip(x_values, poisson_probs):
    print(f'Probability of {k} events: {prob:.4f}')

# Step 5: Plot the Poisson distribution
plt.bar(x_values, poisson_probs, color='purple')
plt.title('Poisson Distribution (λ=3)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(x_values)
plt.show()
```

## Question_6th:- Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?

```python
# Probability Distribution:-
        # Definition: A probability distribution describes how the probabilities of a random variable are distributed across 
          # its possible values. It provides a comprehensive way to represent all the possible outcomes and their associated probabilities.
        

      # Types: There are two main types of probability distributions:
    
          # Discrete Probability Distribution: Deals with discrete random variables (e.g., Binomial, Poisson).
        
          # Continuous Probability Distribution: Deals with continuous random variables (e.g., Normal, Exponential).
            
            
# Probability Density Function (PDF):-

     # Definition: A PDF is a function that describes the likelihood of a continuous random variable taking on a particular value. The PDF itself does not give probabilities directly; instead, the probability of the random variable falling within a certain range is obtained by integrating the PDF over that range.

     # Properties:
        
               # The area under the PDF curve over the entire space equals 1.
            
               # The probability of the random variable being exactly equal to a specific value is always 0.
                
# Probability Mass Function (PMF):-

    # Definition: A PMF is a function that gives the probability of a discrete random variable taking on a specific value. 
      # It provides the probabilities directly for each possible outcome.
        
        
        
        
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom

# Step 1: Parameters for the Binomial distribution
n = 10   # Number of trials
p = 0.5  # Probability of success

# Step 2: Parameters for the Normal distribution
mu = 0   # Mean
sigma = 1  # Standard deviation

# Step 3: Create a range of x values for the PMF (Binomial)
x_binom = np.arange(0, n + 1)

# Step 4: Calculate PMF for Binomial distribution
binomial_probs = binom.pmf(x_binom, n, p)

# Step 5: Create a range of x values for the PDF (Normal)
x_norm = np.linspace(-4, 4, 100)

# Step 6: Calculate PDF for Normal distribution
normal_probs = norm.pdf(x_norm, mu, sigma)

# Step 7: Plotting PMF and PDF
plt.figure(figsize=(12, 6))

# PMF Plot for Binomial Distribution
plt.subplot(1, 2, 1)
plt.bar(x_binom, binomial_probs, color='orange')
plt.title('Binomial PMF (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(x_binom)

# PDF Plot for Normal Distribution
plt.subplot(1, 2, 2)
plt.plot(x_norm, normal_probs, color='blue')
plt.fill_between(x_norm, normal_probs, alpha=0.1)  # Shade the area under the curve
plt.title('Normal PDF (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plots
plt.tight_layout()
plt.show()
```

## Question_7th:- Explain the Central Limit Theorem (CLT) with example.

```python
# The Central Limit Theorem (CLT) is a fundamental statistical principle that states that the distribution of the sample means 
  # (or sums) of a sufficiently large number of independent, identically distributed (i.i.d.) random variables approaches a 
  # normal distribution, regardless of the original distribution of the population.

# Key Points of the Central Limit Theorem:

     # 1.Sample Size: The theorem applies when the sample size n is large (commonly 𝑛≥30).
    
     # 2.Independence: The samples must be independent of each other.
        
     # 3.Normal Approximation: As n increases, the distribution of the sample means will approximate a normal distribution, regardless of the shape of the population distribution.
    
    
    
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define parameters
population_size = 100000
low = 1
high = 10
n_samples = 1000  # Number of samples
sample_size = 30   # Size of each sample

# Step 2: Generate a uniform population
population = np.random.uniform(low, high, population_size)

# Step 3: Take samples and calculate sample means
sample_means = []
for _ in range(n_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Step 4: Plot the population distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(population, bins=30, kde=True, color='blue')
plt.title('Population Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 5: Plot the distribution of sample means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='orange')
plt.title('Distribution of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

## Question_9th:-Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail treject the null hypothesis?

  ## Task: Write Python code to calculate the z-score and p-value for the given data.

  ## Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

```python
import numpy as np
import scipy.stats as stats

# Given data
sample_mean = 105
population_mean = 100
standard_deviation = 15
sample_size = 25


z_score = (sample_mean - population_mean) / (standard_deviation / np.sqrt(sample_size))
p_value = 1 - stats.norm.cdf(z_score)

alpha = 0.05

print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")


if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

## Question_10Th:- Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. Generate 1,000 samples and plot the distribution. What is the expected mean and variance?

   ## Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.

   ## Objective: Understand the properties of a binomial distribution and verify them through simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom


n = 10  
p = 0.6 
num_samples = 1000 

samples = np.random.binomial(n, p, num_samples)


expected_mean = n * p
expected_variance = n * p * (1 - p)

plt.figure(figsize=(10, 6))
sns.histplot(samples, bins=np.arange(-0.5, n + 1.5, 1), kde=False, color='skyblue', stat='density')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Density')
plt.xticks(range(n + 1))
plt.grid(axis='y')
plt.show()

# Output the expected mean and variance
print(f"Expected Mean: {expected_mean:.2f}")
print(f"Expected Variance: {expected_variance:.2f}")
```
